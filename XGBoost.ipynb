```python
import pandas as pd
import xgboost as xg
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset['sex']=le.fit_transform(dataset['sex'])
dataset['smoker']=le.fit_transform(dataset['smoker'])
dataset
```

```python
independent=dataset[['age', 'sex', 'bmi', 'children', 'smoker']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.metrics import mean_squared_error
regressor= xg.XGBRegressor(criterion='squared_error', n_estimators=100)
regressor.fit(x_train, y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test, y_pred)
r_score
```

```python
import pickle
filename="finalized_model_xgboost.sav"
pickle.dump(regressor,open(filename, 'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_xgboost.sav", 'rb'))
result=loaded_model.predict([[45, 0, 40, 2, 0]])
```
